```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mat


N = 6
#values where produced with the script ../scripts/count_retransmissions
values_A = (3.63,2.76,2.64,2.87,3.48,2.89)
values_B = (1.95,1.85,1.95,1.69,1.73,1.62)
mat.rcParams.update({'font.size': 22})
```

```python
mat.rcParams.update({'font.size': 22})
fig, ax = plt.subplots(figsize=(10,6))
index = np.arange(N)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index + 0.15, values_A, bar_width, alpha=opacity, color='red', label = 'A: SP → NY')
rects1 = plt.bar(index + bar_width + 0.15, values_B, bar_width, alpha=opacity, color='orange', label = 'B: NY → SP')

plt.xlabel('Month')
plt.ylabel('TCP retransmission percentage')
strindex = [''] * N
for idx, element in enumerate(index):
    strindex[idx] = (str(element + 3) + '/18')
plt.xticks(index + bar_width, (strindex))
plt.yticks(np.arange(0,4.1,step = 0.5))
leg=plt.legend(loc='upper center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

plt.draw()
bb = leg.get_bbox_to_anchor().inverse_transformed(ax.transAxes)
xOffset = -0.2
bb.x0 += xOffset
bb.x1 += xOffset
leg.set_bbox_to_anchor(bb, transform = ax.transAxes)
plt.show()
```
